```python
# 0. 라이브러리 임포트 및 버전 확인
import tensorflow as tf
import numpy as np
import pandas as pd
import random
import re
import matplotlib.pyplot as plt

from tensorflow.keras.layers import SimpleRNN, LSTM

print("TensorFlow version:", tf.__version__)


# 1. NSMC 데이터 다운로드 및 읽기
#   - 네이버 영화 리뷰 데이터 (train / test) 사용
path_to_train_file = tf.keras.utils.get_file(
    'train.txt',
    'https://raw.githubusercontent.com/e9t/nsmc/master/ratings_train.txt'
)
path_to_test_file = tf.keras.utils.get_file(
    'test.txt',
    'https://raw.githubusercontent.com/e9t/nsmc/master/ratings_test.txt'
)

# 파일 전체 내용을 문자열로 읽기
train_text = open(path_to_train_file, 'rb').read().decode(encoding='utf-8')
test_text = open(path_to_test_file, 'rb').read().decode(encoding='utf-8')

print('Length of train text: {} characters'.format(len(train_text)))
print('Length of test text: {} characters'.format(len(test_text)))
print(train_text[:300])



# 2. 레이블(label) 추출
train_Y = np.array([
    [int(row.split('\t')[2])]
    for row in train_text.split('\n')[1:]
    if row.count('\t') > 0
])

test_Y = np.array([
    [int(row.split('\t')[2])]
    for row in test_text.split('\n')[1:]
    if row.count('\t') > 0
])

print("train_Y shape:", train_Y.shape)
print("test_Y shape:", test_Y.shape)
print("train_Y sample:", train_Y[:5])



# 3. 텍스트 정제 함수 정의
def clean_str(string):
    string = re.sub(r"[^가-힣A-Za-z0-9(),!?\'\`]", " ", string)
    string = re.sub(r"\'s", " \'s", string)
    string = re.sub(r"\'ve", " \'ve", string)
    string = re.sub(r"n\'t", " n\'t", string)
    string = re.sub(r"\'re", " \'re", string)
    string = re.sub(r"\'d", " \'d", string)
    string = re.sub(r"\'ll", " \'ll", string)
    string = re.sub(r",", " , ", string)
    string = re.sub(r"!", " ! ", string)
    string = re.sub(r"\(", " \( ", string)
    string = re.sub(r"\)", " \) ", string)
    string = re.sub(r"\?", " \? ", string)
    string = re.sub(r"\s{2,}", " ", string)
    string = re.sub(r"\'{2,}", "\'", string)
    string = re.sub(r"\'", "", string)
    return string.lower()


# 4. 훈련 텍스트 전처리
train_text_X = [
    row.split('\t')[1]
    for row in train_text.split('\n')[1:]
    if row.count('\t') > 0
]
train_text_X = [clean_str(sentence) for sentence in train_text_X]

sentences = [sentence.split(' ') for sentence in train_text_X]

for i in range(5):
    print(sentences[i])


# 5. TextVectorization으로 정수 인코딩 + 패딩
VOCAB_SIZE = 2000
MAX_LEN = 25

vectorize_layer = tf.keras.layers.TextVectorization(
    standardize='lower_and_strip_punctuation',
    split='whitespace',
    max_tokens=VOCAB_SIZE,
    output_mode='int',
    output_sequence_length=MAX_LEN
)

vectorize_layer.adapt(train_text_X)

train_X = vectorize_layer(train_text_X)
print(train_X[:5])



# 6. 문장 길이 분포 확인 (전처리 후)
sentence_len = [len(sentence) for sentence in sentences]
sentence_len.sort()

plt.plot(sentence_len)
plt.title("Sentence length distribution")
plt.xlabel("Sentence index (sorted)")
plt.ylabel("Length")
plt.show()

print("길이 25 이하 문장 수:", sum([int(l <= 25) for l in sentence_len]))


# 7. 테스트 텍스트 전처리 및 벡터화
test_text_X = [
    row.split('\t')[1]
    for row in test_text.split('\n')[1:]
    if row.count('\t') > 0
]
test_text_X = [clean_str(sentence) for sentence in test_text_X]

test_X = vectorize_layer(test_text_X)

print("test_X shape:", test_X.shape)
print("test_Y shape:", test_Y.shape)



# 8. 공통 하이퍼파라미터 및 랜덤 시드 설정
VOCAB_SIZE = 2000
EMBEDDING_DIM = 128
MAX_LEN = 25
EPOCHS = 10
BATCH_SIZE = 32

random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)


# 9. 기본 Dense 모델 (GlobalAveragePooling 기반)
base_model = tf.keras.Sequential([
    tf.keras.layers.Embedding(VOCAB_SIZE, EMBEDDING_DIM, input_shape=(MAX_LEN,)),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
base_model.summary()

base_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history_base = base_model.fit(
    train_X, train_Y,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=0.2,
    verbose=1
)

# 테스트 데이터에서 성능 평가
test_loss_base, test_acc_base = base_model.evaluate(test_X, test_Y, verbose=0)
print(f"[Base Dense] Test Loss: {test_loss_base:.4f}, Test Accuracy: {test_acc_base:.4f}")


# 10. Dense 모델로 예시 문장 감성분석
example_sentences = [
    "이 영화 진짜 재미있어요",
    "완전 지루하고 별로였음",
    "배우 연기는 좋았지만 스토리가 아쉬웠다"
]

# 예시 문장도 동일하게 정제 후 벡터화
example_sentences_clean = [clean_str(s) for s in example_sentences]
example_seq = vectorize_layer(example_sentences_clean)
pred = base_model.predict(example_seq)

for s, p in zip(example_sentences, pred):
    print(f"문장: {s}")
    print(f"긍정 확률: {p[0]:.4f}")
    print("결과:", "긍정" if p[0] > 0.5 else "부정")


# 11. SimpleRNN 모델
rnn_model = tf.keras.Sequential([
    tf.keras.Input(shape=(MAX_LEN,)),
    tf.keras.layers.Embedding(
        input_dim=VOCAB_SIZE,
        output_dim=EMBEDDING_DIM,
        input_length=MAX_LEN
    ),
    SimpleRNN(units=50),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid'),
])

rnn_model.summary()

rnn_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history_rnn = rnn_model.fit(
    train_X, train_Y,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=0.2,
    verbose=1
)

test_loss_rnn, test_acc_rnn = rnn_model.evaluate(test_X, test_Y, verbose=0)
print(f"[SimpleRNN] Test Loss: {test_loss_rnn:.4f}, Test Accuracy: {test_acc_rnn:.4f}")


# 12. SimpleRNN: 단어 순서에 따른 차이 확인 예시
example_sentences = [
    "배우 연기는 좋았지만 스토리가 아쉬웠다",
    "스토리는 아쉬웠지만 배우 연기는 좋았다"
]

example_sentences_clean = [clean_str(s) for s in example_sentences]
example_seq = vectorize_layer(example_sentences_clean)
pred = rnn_model.predict(example_seq)

print("\n[SimpleRNN] 순서에 따른 예측 차이")
for s, p in zip(example_sentences, pred):
    print(f"문장: {s}")
    print(f"긍정 확률: {p[0]:.4f}")
    print("결과:", "긍정" if p[0] > 0.5 else "부정")


# 13. Dense vs SimpleRNN 비교 예시
example_sentences = [
    "배우연기는 좋았지만 스토리가 아쉬웠다",
    "스토리는 아쉬웠지만 배우 연기는 좋았다"
]

example_sentences_clean = [clean_str(s) for s in example_sentences]
example_seq = vectorize_layer(example_sentences_clean)
pred_dense = base_model.predict(example_seq)

print("\n[Base Dense] 순서에 따른 예측")
for s, p in zip(example_sentences, pred_dense):
    print(f"문장: {s}")
    print(f"긍정 확률: {p[0]:.4f}")
    print("결과:", "긍정" if p[0] > 0.5 else "부정")


# 14. LSTM 모델
lstm_model = tf.keras.Sequential([
    tf.keras.Input(shape=(MAX_LEN,)),
    tf.keras.layers.Embedding(
        input_dim=VOCAB_SIZE,
        output_dim=EMBEDDING_DIM,
        input_length=MAX_LEN
    ),
    LSTM(units=50),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid'),
])

lstm_model.summary()

lstm_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history_lstm = lstm_model.fit(
    train_X, train_Y,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=0.2,
    verbose=1
)

# lstm_model.evaluate(test_X, test_Y) 결과 출력
test_loss_lstm, test_acc_lstm = lstm_model.evaluate(test_X, test_Y, verbose=0)
print(f"[LSTM] Test Loss: {test_loss_lstm:.4f}, Test Accuracy: {test_acc_lstm:.4f}")


# 15. LSTM으로 긴 예시 문장 감성분석
example_sentences = [
    "이 영화는 정말 재미있고, 스토리도 흥미진진하며 배우들의 연기까지 완벽해서 끝까지 눈을 뗄 수 없었었요.",
    "영화가 전체적으로 너무 지루하고 전개가 느려서 몰입하기 힘들었으며, 기대했던 재미나 감동이 전혀 없었어요."
]

example_sentences_clean = [clean_str(s) for s in example_sentences]
example_seq = vectorize_layer(example_sentences_clean)
pred_lstm = lstm_model.predict(example_seq)

print("\n[LSTM] 긴 문장 예측 결과")
for s, p in zip(example_sentences, pred_lstm):
    print(f"문장: {s}")
    print(f"긍정 확률: {p[0]:.4f}")
    print("결과:", "긍정" if p[0] > 0.5 else "부정")


# 16. SimpleRNN과 LSTM 결과 비교 (같은 문장)
pred_rnn = rnn_model.predict(example_seq)

print("\n[SimpleRNN] 동일 문장 예측 결과")
for s, p in zip(example_sentences, pred_rnn):
    print(f"문장: {s}")
    print(f"긍정 확률: {p[0]:.4f}")
    print("결과:", "긍정" if p[0] > 0.5 else "부정")
```
